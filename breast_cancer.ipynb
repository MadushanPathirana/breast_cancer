# Description 

Missing Attribute Values: (denoted by "?")

"Class" variable typically refers to whether a patient experienced a recurrence of breast cancer after initial treatment. The two possible values for this variable are:

"no-recurrence-events": This indicates that the patient did not experience a recurrence of breast cancer during the follow-up period.
"recurrence-events": This indicates that the patient did experience a recurrence of breast cancer during the follow-up period.

"menopause" refers to the age at which a woman experienced menopause. It can be categorized as:

"lt40" (less than 40 years old): Women who experienced menopause before the age of 40.
"ge40" (greater than or equal to 40 years old): Women who experienced menopause at the age of 40 or later.
"premeno" (premenopausal): Women who have not yet experienced menopause.

"Invasive nodes" (abbreviated as "inv_nodes" in some medical contexts) refers to the number of lymph nodes that are affected by cancer in cases of invasive breast cancer.

"Node caps" (abbreviated as "node_caps" or "caps" in some medical contexts) refers to whether cancer has spread beyond the wall of a lymph node and into the surrounding tissues, as detected by a pathologist during examination of the lymph node under a microscope.

"Deg_malig" (or "degree of malignancy") refers to the degree or extent to which cancer cells are abnormal or different from normal cells, as determined by a pathologist during examination of the tumor tissue under a microscope.

"Irradiat" (or "irradiation") is a term used in the medical context of breast cancer treatment to refer to the use of radiation therapy to kill cancer cells or prevent them from growing.
